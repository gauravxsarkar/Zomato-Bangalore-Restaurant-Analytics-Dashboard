```python
single_restaurant_types = df[~df['rest_type'].str.contains(',')]['rest_type'].value_counts()
single_restaurant_types
```

```python
# Function that handles ANY number of combined types
def assign_single_type(rest_type_str, single_counts):
    # If already a single type, return as-is
    if ',' not in str(rest_type_str):
        return rest_type_str.strip()
    
    # Split into individual types (works for 2, 3, or more)
    types = [t.strip() for t in rest_type_str.split(',')]
    
    # Keep only types that exist in our single-type reference
    valid_types = [t for t in types if t in single_counts.index]
    
    if not valid_types:
        return 'Other'  # fallback: just take the first one
    
    # Weighted probability based on real counts
    weights = np.array([single_counts[t] for t in valid_types], dtype=float)
    weights = weights / weights.sum()  # normalize to sum to 1
    
    return np.random.choice(valid_types, p=weights)

# Apply to entire dataset
np.random.seed(42)
df['rest_type_clean'] = df['rest_type'].apply(
    lambda x: assign_single_type(x, single_restaurant_types)
)
```

```python
df['rest_type_clean'] = df['rest_type_clean'].replace(['Dhaba','Food Truck','Kiosk','Confectionery','Mess','Takeaway','Sweet Shop','Beverage Shop','Food Court','Bhojanalya','Pop Up'],'Other')
```

```python
# grouping high revenue generating groups together
df['rest_type_clean'] = df['rest_type_clean'].replace(['Bar','Pub','Lounge','Fine Dining','Microbrewery','Club'],'Fine dining/Clubs')
```

**cleaning location coln**

```python
location = df['location'].value_counts(ascending=False)
```

```python
location_less_than_200 = location[location<200]

def handle_location(value):
    if(value in location_less_than_200):
        return 'Others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts()
cuisines_less_than_100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_less_than_100):
        return 'Others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
